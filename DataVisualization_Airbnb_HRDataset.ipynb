```python
import pandas as pd
import seaborn as sns
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
airbnb = pd.read_csv('AirbnbDataset.csv', encoding = ('ISO-8859-1'))
```

```python
area = airbnb['neighbourhood_group']
price = airbnb['price']
fig = plt.figure(figsize=(10,5))
plt.bar(area, price, color = 'maroon', width = 0.4)
plt.xlabel('Area')

plt.ylabel('Price')
plt.title('Area by Price')
plt.show()
```

The columns using to visualize are 'neighbourhood_group' and 'price' from 'Airbnb dataset' with 'Bar chart' representation. From this above chart we can analyze that variation in price depends on area by area.

```python
review = airbnb['number_of_reviews']
price = airbnb['price']
fig = plt.figure(figsize=(10,5))
plt.scatter(price, review, color = 'maroon')
plt.xlabel('Price')
plt.ylabel('Number of Reviews')
plt.title('Number of Reviews by Price')
plt.show()
```

The columns using to visualize are 'number_of_reviews' and 'price' from 'Airbnb dataset' with 'Scatter chart' representation. From this above chart we can analyze that most number of people like to stay in less price and their reviews are higher in those areas.

```python
fig = plt.figure(figsize=(10,5))
print(airbnb.columns)
corr = airbnb.corr( method = 'kendall')
sns.heatmap(corr, annot = True)
plt.title('Correlation between different features')
plt.show()
```

Here using all the columns for visualization from 'Airbnb dataset' with 'Heatmap' representation. From this above visualization we have seen all the correlation between different features or variables.

```python
fig = plt.figure(figsize=(10,5))
ax = sns.countplot(y = 'room_type', hue = 'neighbourhood_group', data = airbnb, palette = 'bright')

total = len(airbnb['room_type'])
for i in ax.patches:
    percentage = "{:.1f}%".format(100*i.get_width()/total)
    x = i.get_x() + i.get_width() + 0.02
    y = i.get_y() + i.get_height()/2
    ax.annotate(percentage,(x,y))
    
plt.xlabel('Room Count')
plt.xticks(rotation = 90)
plt.ylabel('Room Types')
plt.title('Count of each room types')
plt.show()
```

The columns using to visualize are 'neighbourhood_group' and 'room_type' from 'Airbnb dataset' with 'countplot chart' representation. From this above chart we can analyze that different types of rooms count grouped by neighbourhood_group.

```python
fig = plt.figure(figsize=(10,5))
plt.hist(airbnb['neighbourhood'], bins = 20, color= 'maroon')
plt.xlabel('Neighbourhood')
plt.xticks(rotation = 90)
plt.ylabel('Frequency')
plt.title('Histogram of Neighbourhood')
plt.show()
```

The column using to visualize is 'neighbourhood' from 'Airbnb dataset' with 'Histogram chart' representation. From this above chart we can analyze that the frequency or the occurance of neighbourhood.

```python
hr_dataset = pd.read_csv('HRDataset_v14.csv')
```

```python
fig = plt.figure(figsize=(10,5))
sns.boxplot(hr_dataset['PerformanceScore'],hr_dataset['Salary'])
plt.title('Performance Score by Salary')
plt.xticks(rotation = 90)
plt.show()
```

The column using to visualize are 'Performance Score' and 'Salary' from 'HRDataset' with 'boxplot chart' representation. From this above chart we can analyze that the Employees having salary according to their performance score.

```python
fig = plt.figure(figsize=(10,5))
sns.distplot(hr_dataset['Salary'])
plt.title('Salary of Employees')
plt.show()
```

The column using to visualize is 'Salary' from 'HRDataset' with 'Distplot chart' representation. From this above chart we can analyze that the maximum number of Employees have salary range between 50 thousand to 1 lakhs.

```python
fig = plt.figure(figsize=(10,5))
sns.stripplot(hr_dataset['MaritalDesc'],hr_dataset['Salary'], jitter = True)
plt.title('Marital Status by Salary')
plt.show()
```

The column using to visualize are 'MaritalDesc' and 'Salary' from 'HRDataset' with 'Stripplot chart' representation. From this above chart we can analyze that the Salary distribution among the Marital Status.

```python
fig = plt.figure(figsize=(10,5))
sns.violinplot(hr_dataset['Department'],hr_dataset['Absences'])
plt.title('Number of Employees absent in each department')
plt.show()
```

The column using to visualize are 'Department' and 'Absences' from 'HRDataset' with 'violinplot chart' representation. From this above chart we can analyze that the number of Employees absent in each department.

```python
data = hr_dataset['Department'].value_counts()
plt.pie(data, labels = data.index, autopct = '%1.1f%%')
plt.title('Distribution of Employees by Department')
plt.show()
```

The column using to visualize is 'Department' from 'HRDataset' with 'Pie chart' representation. From this above chart we can analyze that the number of Employees in each department.
